```python
import pandas as pd
```

```python
df = pd.read_csv('book.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.strip()
```

```python
df.dtypes
```

```python
df["Total Revenue (in M)"] = (
    df["Total Revenue (in M)"]
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
df["Net Income (in M)"] = (
    df["Net Income (in M)"]
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
df
```

```python
df["Revenue Growth (%)"] = (
    df.groupby("Company")["Total Revenue (in M)"]
    .pct_change(fill_method=None) * 100
)
```

```python
df["Net Income Growth (%)"] = (
    df.groupby("Company")["Net Income (in M)"]
    .pct_change(fill_method=None) * 100
)
```

```python
df
```

```python
df.groupby("Company")["Total Revenue (in M)"].mean()
```

```python
df.groupby("Fiscal Year")["Total Revenue (in M)"].sum()
```

```python
df.groupby("Company")["Revenue Growth (%)"].mean()
```
