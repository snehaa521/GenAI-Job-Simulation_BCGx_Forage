```python
def simple_chatbot(user_query):
    
    if user_query.lower() == "what is microsoft's revenue growth?":
        return "Microsoft's revenue declined by approximately 13%."

    elif user_query.lower() == "how has tesla's net income changed?":
        return "Tesla's net income increased significantly, with growth above 85% in the recent year."

    elif user_query.lower() == "which company has the highest average revenue?":
        return "Apple has the highest average revenue among the three companies."

    elif user_query.lower() == "is apple's revenue increasing or decreasing?":
        return "Apple's revenue has slightly decreased in recent years."

    else:
        return "Sorry, I can only answer predefined financial questions."
```

```python
while True:
    user_input = input("Ask a financial question (type 'exit' to stop): ")
    
    if user_input.lower() == "exit":
        print("Chatbot ended.")
        break
    
    response = simple_chatbot(user_input)
    print(response)
```

Chatbot Prototype

. Predefined Queries
What is Microsoft's revenue growth?
How has Tesla's net income changed?

. Limitations
The chatbot cannot answer dynamic or new questions.
It only responds to exact predefined queries.
